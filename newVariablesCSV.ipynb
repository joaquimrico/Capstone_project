```python
import pandas as pd
import numpy as np

df = pd.read_csv("Dades_meteorol_giques_de_la_XEMA.csv",low_memory=False)
```

```python
rainingDays = df[df['VALOR_LECTURA']!=0]
```

```python
final_columns = ['DATA_LECTURA','VALOR_LECTURA']

rainingDays = rainingDays[final_columns]
```

```python
rainingDays['datetime'] = pd.to_datetime(rainingDays['DATA_LECTURA'])
    
# extract day, month, year, hour, and minute components and store in new columns
rainingDays['day'] = rainingDays['datetime'].dt.day
rainingDays['month'] = rainingDays['datetime'].dt.month
rainingDays['year'] = rainingDays['datetime'].dt.year
rainingDays['hour'] = rainingDays['datetime'].dt.hour

# group by 

grouping_columns = ['year','month','day','hour']

rainingDaysG = rainingDays.groupby(grouping_columns)['VALOR_LECTURA'].mean().reset_index()

# we sort the data 
rainingDaysG = rainingDaysG.sort_values(by=['year','month','day','hour'])
# shouw rainingDaysG
rainingDaysG.head(10)
```

```python
festive_days = ['2022-01-01','2022-01-06','2022-04-15','2022-04-18','2022-06-06','2022-06-24','2022-08-15','2022-09-24','2022-09-26', '2022-10-12','2022-11-01','2022-12-06','2022-12-08','2022-12-26',
                       '2023-01-06','2023-04-07','2023-04-10','2023-05-01','2023-06-05','2023-06-24','2023-08-15','2023-09-11', '2023-09-25','2023-10-12','2023-11-01','2023-12-06','2023-12-08','2023-12-25','2023-12-25','2023-12-26',
                       '2021-01-01','2021-01-06','2021-04-02','2021-04-05','2021-05-01','2021-06-24','2021-09-11','2021-10-12', '2021-11-01','2021-12-06','2021-12-08','2021-12-25']

festiveDays = pd.DataFrame(data=festive_days)
festiveDays.columns = ['datetime']

import datetime as dt

festiveDays['datetime'] = pd.to_datetime(festiveDays['datetime'],
                                        format='%Y-%m-%d')

# extract day, month, year, hour, and minute components and store in new columns
festiveDays['day'] = festiveDays['datetime'].dt.day
festiveDays['month'] = festiveDays['datetime'].dt.month
festiveDays['year'] = festiveDays['datetime'].dt.year
festiveDays.drop(columns=['datetime'])

festiveDays['holiday'] = True
festiveDays.head(10)
```

```python
from datetime import timedelta,datetime

start_dt = datetime(2021, 1, 1,0,0,0,0)
end_dt = datetime(2024, 1, 1,0,0,0,0)

# difference between current and previous date
delta = timedelta(hours=1)

# store the dates between two dates in a list
dates = []

while start_dt <= end_dt:
    # add current date to list by converting  it to iso format
    dates.append(start_dt)
    # increment start date by timedelta
    start_dt += delta
```

```python
all_dates = pd.DataFrame(data=dates)
all_dates.columns = ['datetime']
```

```python
import datetime as dt
all_dates['datetime'] = pd.to_datetime(all_dates['datetime'])

# extract day, month, year, hour, and minute components and store in new columns
all_dates['day'] = all_dates['datetime'].dt.day
all_dates['month'] = all_dates['datetime'].dt.month
all_dates['year'] = all_dates['datetime'].dt.year
all_dates['hour'] = all_dates['datetime'].dt.hour
all_dates.drop(columns=['datetime'])
```

```python
merge_df = pd.merge(all_dates[['day','month','year','hour']],rainingDaysG,
                how='left',
                left_on=['day','month','year','hour'],
                right_on=['day','month','year','hour'])

final_df = pd.merge(merge_df,festiveDays[['day','month','year','holiday']],
                how='left',
                left_on=['day','month','year'],
                right_on=['day','month','year'])

print(final_df.shape)
print(final_df.head(25))
```

```python
import numpy as np
final_df['VALOR_LECTURA'] = final_df['VALOR_LECTURA'].fillna(0)
final_df['holiday'] = final_df['holiday'].replace(np.nan, False)
```

```python
final_df.to_csv("newColumns.csv",index=False)
```
